```python
database_coap_fields            = ["coap_type", "coap_code", "coap_retransmitted"]
database_http_fields            = ["http_request", "http_request_method", "http_response_code"]
database_message_pattern_fields = ["message_protocol", "message_size", "message_source", "message_destination"]
```

```python
%%time

coap_protocol_df = (
  real_df
  .select(database_coap_fields)
  .unique()
  .drop_nulls()
)

http_protocol_df = (
  real_df
  .select(database_http_fields)
  .unique()
)

protocols_df = (
  real_df
  .select(["message_protocol"] + database_coap_fields + database_http_fields)
  .unique()
)

message_patterns_df = (
  real_df
  .select(database_message_pattern_fields + database_coap_fields + database_http_fields)
  .unique()
)

message_patterns_df
```

```python
# Drop all-null rows
http_protocol_df = http_protocol_df.filter(
    ~pl.fold(
        acc=True,
        f=lambda acc, s: acc & s.is_null(),
        exprs=pl.all(),
    )
)

http_protocol_df
```

```python
import dateutil.parser as dp
import pandas as pd

F = "/home/ubuntu/dos-californium/deter/expdata/real/final/{}/{}/client1.log"

d = list()
def parse_client_log(F, expname, trial):
  with open(F) as f:
    for i, L in enumerate(f):
      if i == 0:
        continue

      L = L.replace("--", "")
      L = L.replace("d:", "d ")
      L = L.replace("v:", "v ")

      parts = L.split()
      parts = [f"{parts[0]} {parts[1]}", parts[2].lower(), int(parts[3])]
      parts[0] = dp.parse(parts[0]).timestamp()
      
      d.append({
        "exp_id": expname,
        "trial": trial,
        "timestamp": parts[0],
        "event": parts[1],
        "message_number": parts[2],
      })
    
exps = [
  "gigabit_attack_50sec_with_tcpdump",
  "gigabit_attack_50sec_without_tcpdump",
  "gigabit_attack_100sec_with_tcpdump",
  "gigabit_attack_100sec_without_tcpdump",

]
trials = list(range(1, 3 + 1))
for e in exps:
  for t in trials:
    parse_client_log(F.format(e, t), e, t)
    
df = pd.DataFrame.from_dict(d)

df.loc[df["event"] == "send", "timestamp"] *= -1

df
```

```python
tdf = (
  df.groupby(by=["exp_id", "trial", "message_number"])
  .agg(
    rtt_sec=("timestamp", "sum")
  )
  .reset_index()
)

tdf["rtt_ms"] = (1e3 * tdf["rtt_sec"]).astype(int)
tdf.drop(columns=["rtt_sec"], inplace=True)

tdf = tdf[tdf["rtt_ms"] > 0]

tdf.loc[tdf["exp_id"] == "gigabit_attack_50sec_with_tcpdump", "attack_duration"] = 50
tdf.loc[tdf["exp_id"] == "gigabit_attack_50sec_without_tcpdump", "attack_duration"] = 50
tdf.loc[tdf["exp_id"] == "gigabit_attack_100sec_with_tcpdump", "attack_duration"] = 100
tdf.loc[tdf["exp_id"] == "gigabit_attack_100sec_without_tcpdump", "attack_duration"] = 100

tdf.loc[tdf["exp_id"] == "gigabit_attack_50sec_with_tcpdump", "run_tcpdump"] = True
tdf.loc[tdf["exp_id"] == "gigabit_attack_50sec_without_tcpdump", "run_tcpdump"] = False
tdf.loc[tdf["exp_id"] == "gigabit_attack_100sec_with_tcpdump", "run_tcpdump"] = True
tdf.loc[tdf["exp_id"] == "gigabit_attack_100sec_without_tcpdump", "run_tcpdump"] = False

tdf
```

```python
xdf = (
  tdf.groupby(by=["exp_id", "trial"])
  .agg(
    num_messages=("message_number", "max"),
    avg_rtt_ms=("rtt_ms", "mean"),
    max_rtt_ms=("rtt_ms", "max"),
  )
  .reset_index()
)

xdf
```

```python
import pandas as pd

regdf = pd.read_csv("/home/ubuntu/dos-californium/regdumps.csv")
regdf["zipped"] = False
zipdf = pd.read_csv("/home/ubuntu/dos-californium/zipdumps.csv")
zipdf["zipped"] = True

_df = pd.concat([regdf, zipdf])
_df
d = list()
for r in _df.to_dict(orient="records"):
  nr = {**r}
  parts = r["exp_id"].split("/")
  nr["exp_id"] = parts[0]
  nr["trial"] = parts[1]
  nr["dump"] = parts[2].replace("client1", "client").replace(".zip", "").replace(".pcap", "")
  d.append(nr)
  
df = pd.DataFrame.from_records(d)

gdf = (
  df
  .groupby(by=["exp_id", "trial", "dump"])
  .agg(
    unzipped_size=("size", "max"),
    zipped_size=("size", "min"),
  )
  .reset_index()
)
gdf["compression_ratio"] = gdf["unzipped_size"] / gdf["zipped_size"]

import polars as pl

cdf = pl.read_parquet("/home/ubuntu/cloudgroupe2e/configurations.parquet").to_pandas()
cdf = cdf[["exp_id", "attacker_rate", "run_proxy_with_dtls", "run_proxy_with_https", "run_attacker"]]


mdf = gdf.merge(cdf, on="exp_id")

mdf["dump_distribution"] = mdf["dump"]
mdf.dump_distribution.replace({
  "attacker_dump": "CoAP Dominant",
  "client_dump": "CoAP Dominant",
  "proxy_dump": "Mixed",
  "receiver_dump": "CoAP Dominant",
  "server_dump": "HTTP Dominant",
}, inplace=True)

mdf.rename(columns={
  "run_proxy_with_dtls": "dtls",
  "run_proxy_with_https": "https",
  "dump_distribution": "capture_distribution",
}, inplace=True)

mdf["encryption"] = ""
mdf.loc[(mdf["dtls"] == False) & (mdf["https"] == False), "encryption"] = "No Encryption"
mdf.loc[(mdf["dtls"] == False) & (mdf["https"] == True), "encryption"] = "HTTPS"
mdf.loc[(mdf["dtls"] == True) & (mdf["https"] == False), "encryption"] = "CoAPS"
mdf.loc[(mdf["dtls"] == True) & (mdf["https"] == True), "encryption"] = "CoAPS + HTTPS"

import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300

g = sns.catplot(
  data=mdf,
  kind="bar",
  ci=None,
  x="encryption",
  y="compression_ratio",
  order=["No Encryption", "CoAPS", "HTTPS", "CoAPS + HTTPS"],
  hue="capture_distribution",
  hue_order=["CoAP Dominant", "HTTP Dominant", "Mixed"],
  legend_out=False,
  height=5,
  aspect=2,
)

plt.rcParams['font.size'] = '18'
plt.rcParams['lines.linewidth'] = 3
plt.rcParams['lines.markersize'] = 9

g.set(xlabel="Encryption Type", ylabel="Average Compression Ratio")
sns.move_legend(
  g, 
  "upper right", 
  title="Traffic Type", 
  frameon=True, 
  labelspacing=0.05,
  bbox_to_anchor=((1, 1)),
  markerfirst=True,
)
plt.show()

mdf
```
