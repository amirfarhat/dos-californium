```python
%%time

import polars as pl

FP = "/home/ubuntu/multi.parquet"
lazy_df = pl.scan_parquet(FP)
real_df = pl.read_parquet(FP)
real_df
```

```python
database_coap_fields            = ["coap_type", "coap_code", "coap_retransmitted"]
database_http_fields            = ["http_request", "http_request_method", "http_response_code"]
database_message_pattern_fields = ["message_protocol", "message_size", "message_source", "message_destination"]
```

```python
%%time

coap_protocol_df = (
  real_df
  .select(database_coap_fields)
  .unique()
  .drop_nulls()
)

http_protocol_df = (
  real_df
  .select(database_http_fields)
  .unique()
)

protocols_df = (
  real_df
  .select(["message_protocol"] + database_coap_fields + database_http_fields)
  .unique()
)

message_patterns_df = (
  real_df
  .select(database_message_pattern_fields + database_coap_fields + database_http_fields)
  .unique()
)

message_patterns_df
```

```python
coap_protocol_df
```

```python
# Drop all-null rows
http_protocol_df = http_protocol_df.filter(
    ~pl.fold(
        acc=True,
        f=lambda acc, s: acc & s.is_null(),
        exprs=pl.all(),
    )
)

http_protocol_df
```

```python
%%time

# print(f"{len(protocols_df)}")
# print(f"{len(coap_protocol_df)}")
# print(f"{len(http_protocol_df)}")

assert len(protocols_df) == len(http_protocol_df) + len(coap_protocol_df)

unique_coap_df = (
  protocols_df
  .filter(pl.col("message_protocol") == "coap")
  .select(["coap_type", "coap_code", "coap_retransmitted"])
  .unique()
)

unique_http_df = (
  protocols_df
  .filter(pl.col("message_protocol") == "http")
  .select(["http_request", "http_request_method", "http_response_code"])
  .unique()
)

mp_coap_df = (
  message_patterns_df
  .filter(pl.col("message_protocol") == "coap")
  .select(database_coap_fields)
  .unique()
)

mp_http_df = (
  message_patterns_df
  .filter(pl.col("message_protocol") == "http")
  .select(database_http_fields)
  .unique()
)

assert unique_coap_df.frame_equal(coap_protocol_df)
assert mp_coap_df.frame_equal(coap_protocol_df)

assert unique_http_df.frame_equal(http_protocol_df)
assert mp_http_df.frame_equal(http_protocol_df)
```
