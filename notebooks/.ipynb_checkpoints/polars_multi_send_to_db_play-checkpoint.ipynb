```python
%%time

import polars as pl

FP = "/home/ubuntu/multi.parquet"
lazy_df = pl.scan_parquet(FP)
real_df = pl.read_parquet(FP)
real_df
```

```python
database_coap_fields            = ["coap_type", "coap_code", "coap_retransmitted"]
database_http_fields            = ["http_request", "http_request_method", "http_response_code"]
database_message_pattern_fields = ["message_protocol", "message_size", "message_source", "message_destination"]
```

```python
%%time

coap_protocol_df = (
  real_df
  .select(database_coap_fields)
  .unique()
  .drop_nulls()
)

http_protocol_df = (
  real_df
  .select(database_http_fields)
  .unique()
)

protocols_df = (
  real_df
  .select(["message_protocol"] + database_coap_fields + database_http_fields)
  .unique()
)

message_patterns_df = (
  real_df
  .select(database_message_pattern_fields + database_coap_fields + database_http_fields)
  .unique()
)

message_patterns_df
```

```python
coap_protocol_df
```

```python
# Drop all-null rows
http_protocol_df = http_protocol_df.filter(
    ~pl.fold(
        acc=True,
        f=lambda acc, s: acc & s.is_null(),
        exprs=pl.all(),
    )
)

http_protocol_df
```

```python
%%time

# print(f"{len(protocols_df)}")
# print(f"{len(coap_protocol_df)}")
# print(f"{len(http_protocol_df)}")

assert len(protocols_df) == len(http_protocol_df) + len(coap_protocol_df)

unique_coap_df = (
  protocols_df
  .filter(pl.col("message_protocol") == "coap")
  .select(["coap_type", "coap_code", "coap_retransmitted"])
  .unique()
)

unique_http_df = (
  protocols_df
  .filter(pl.col("message_protocol") == "http")
  .select(["http_request", "http_request_method", "http_response_code"])
  .unique()
)

mp_coap_df = (
  message_patterns_df
  .filter(pl.col("message_protocol") == "coap")
  .select(database_coap_fields)
  .unique()
)

mp_http_df = (
  message_patterns_df
  .filter(pl.col("message_protocol") == "http")
  .select(database_http_fields)
  .unique()
)

assert unique_coap_df.frame_equal(coap_protocol_df)
assert mp_coap_df.frame_equal(coap_protocol_df)

assert unique_http_df.frame_equal(http_protocol_df)
assert mp_http_df.frame_equal(http_protocol_df)
```

```python
import pandas as pd

df = (
  pd.read_json("/home/ubuntu/dump_times.json", orient="index")
  .reset_index()
  .rename(columns={"index": "exp_id"})
  [["exp_id", "attacker_dump.pcap", "client1_dump.pcap", "receiver_dump.pcap", "server_dump.pcap", "proxy_dump.pcap"]]
)

dumps = [c for c in df.columns if "_dump" in c]
for d in dumps:
  df["rescale_" + d] = round(100 * df[d] / df["proxy_dump.pcap"])
  
df["intensity"] = None
df.loc[df["proxy_dump.pcap"] > 200, "intensity"] = "high"
df.loc[(50 <= df["proxy_dump.pcap"]) & (df["proxy_dump.pcap"] <= 200), "intensity"] = "medium"
df.loc[(df["proxy_dump.pcap"] < 50), "intensity"] = "low"

df
```

```python
gdf = (
  df.groupby("intensity")
  .agg(
    avg_proxy_decode_time=("proxy_dump.pcap", "mean"),
    avg_server_decode_time=("server_dump.pcap", "mean"),
    avg_receiver_decode_time=("receiver_dump.pcap", "mean"),
    avg_client1_decode_time=("client1_dump.pcap", "mean"),
    avg_attacker_decode_time=("attacker_dump.pcap", "mean"),
  )
  .reset_index()
)

for c in set(gdf.columns) - {"intensity"}:
  gdf[c] = round(gdf[c])

gdf = gdf[["intensity", "avg_proxy_decode_time", "avg_server_decode_time", "avg_attacker_decode_time", "avg_receiver_decode_time", "avg_client1_decode_time"]]

gdf.to_csv("/home/ubuntu/dump_sizes.csv", index=None)

gdf
```

```python
rdf = gdf.copy()

for c in set(rdf.columns) - {"intensity", "avg_proxy_decode_time"}:
  rdf[c] = 100 * rdf[c] / rdf["avg_proxy_decode_time"]
rdf["avg_proxy_decode_time"] = 100

rdf
```

```python
import polars as pl

F = "/home/ubuntu/gigabit_attack_*.csv"
df = pl.read_csv(F)
df
```

```python
tdf = df.with_columns([
  pl.when(pl.col("exp_id").str.contains("100"))
    .then(pl.lit(100).alias("attack_duration_sec"))
    .otherwise(pl.lit(50).alias("attack_duration_sec")),
  pl.when(pl.col("exp_id").str.contains("without"))
    .then(pl.lit(False).alias("run_tcpdump"))
    .otherwise(pl.lit(True).alias("run_tcpdump")),
])

tdf = tdf.with_columns([
  pl.when(pl.col("run_tcpdump") == False)
    .then((pl.col("p95_metric_value") * -1).alias("p95_metric_value"))
    .otherwise(pl.col("p95_metric_value").alias("p95_metric_value")),
    pl.when(pl.col("run_tcpdump") == False)
    .then((pl.col("p99_metric_value") * -1).alias("p99_metric_value"))
    .otherwise(pl.col("p99_metric_value").alias("p99_metric_value")),
    pl.when(pl.col("run_tcpdump") == False)
    .then((pl.col("mean_metric_value") * -1).alias("mean_metric_value"))
    .otherwise(pl.col("mean_metric_value").alias("mean_metric_value")),
])

tdf
```

```python
(
  tdf.groupby(by=["attack_duration_sec", "node_type", "metric_type"])
  .agg([
    pl.col("p95_metric_value").sum(),
    pl.col("p99_metric_value").sum(),
    pl.col("mean_metric_value").sum(),
  ])
  .sort(by=["attack_duration_sec", "node_type", "metric_type"])
)
```

```python
import dateutil.parser as dp
import pandas as pd

F = "/home/ubuntu/dos-californium/deter/expdata/real/final/{}/{}/client1.log"

d = list()
def parse_client_log(F, expname, trial):
  with open(F) as f:
    for i, L in enumerate(f):
      if i == 0:
        continue

      L = L.replace("--", "")
      L = L.replace("d:", "d ")
      L = L.replace("v:", "v ")

      parts = L.split()
      parts = [f"{parts[0]} {parts[1]}", parts[2].lower(), int(parts[3])]
      parts[0] = dp.parse(parts[0]).timestamp()
      
      d.append({
        "exp_id": expname,
        "trial": trial,
        "timestamp": parts[0],
        "event": parts[1],
        "message_number": parts[2],
      })
    
exps = [
  "gigabit_attack_50sec_with_tcpdump",
  "gigabit_attack_50sec_without_tcpdump",
  "gigabit_attack_100sec_with_tcpdump",
  "gigabit_attack_100sec_without_tcpdump",

]
trials = list(range(1, 3 + 1))
for e in exps:
  for t in trials:
    parse_client_log(F.format(e, t), e, t)
    
df = pd.DataFrame.from_dict(d)

df.loc[df["event"] == "send", "timestamp"] *= -1

df
```

```python
tdf = (
  df.groupby(by=["exp_id", "trial", "message_number"])
  .agg(
    rtt_sec=("timestamp", "sum")
  )
  .reset_index()
)

tdf["rtt_ms"] = (1e3 * tdf["rtt_sec"]).astype(int)
tdf.drop(columns=["rtt_sec"], inplace=True)

tdf = tdf[tdf["rtt_ms"] > 0]

tdf.loc[tdf["exp_id"] == "gigabit_attack_50sec_with_tcpdump", "attack_duration"] = 50
tdf.loc[tdf["exp_id"] == "gigabit_attack_50sec_without_tcpdump", "attack_duration"] = 50
tdf.loc[tdf["exp_id"] == "gigabit_attack_100sec_with_tcpdump", "attack_duration"] = 100
tdf.loc[tdf["exp_id"] == "gigabit_attack_100sec_without_tcpdump", "attack_duration"] = 100

tdf.loc[tdf["exp_id"] == "gigabit_attack_50sec_with_tcpdump", "run_tcpdump"] = True
tdf.loc[tdf["exp_id"] == "gigabit_attack_50sec_without_tcpdump", "run_tcpdump"] = False
tdf.loc[tdf["exp_id"] == "gigabit_attack_100sec_with_tcpdump", "run_tcpdump"] = True
tdf.loc[tdf["exp_id"] == "gigabit_attack_100sec_without_tcpdump", "run_tcpdump"] = False

tdf
```

```python
xdf = (
  tdf.groupby(by=["exp_id", "trial"])
  .agg(
    num_messages=("message_number", "max"),
    avg_rtt_ms=("rtt_ms", "mean"),
    max_rtt_ms=("rtt_ms", "max"),
  )
  .reset_index()
)

xdf
```

```python
(
   tdf.groupby(by=["attack_duration", "run_tcpdump"])
  .agg(
    num_messages=("message_number", "max"),
    avg_rtt_ms=("rtt_ms", "mean"),
    std_rtt_ms=("rtt_ms", "std"),
    max_rtt_ms=("rtt_ms", "max"),
  )
)
```

```python
import pandas as pd

regdf = pd.read_csv("/home/ubuntu/dos-californium/regdumps.csv")
regdf["zipped"] = False
zipdf = pd.read_csv("/home/ubuntu/dos-californium/zipdumps.csv")
zipdf["zipped"] = True

_df = pd.concat([regdf, zipdf])
_df
d = list()
for r in _df.to_dict(orient="records"):
  nr = {**r}
  parts = r["exp_id"].split("/")
  nr["exp_id"] = parts[0]
  nr["trial"] = parts[1]
  nr["dump"] = parts[2].replace("client1", "client").replace(".zip", "").replace(".pcap", "")
  d.append(nr)
  
df = pd.DataFrame.from_records(d)
df
```

```python
gdf = (
  df
  .groupby(by=["exp_id", "trial", "dump"])
  .agg(
    unzipped_size=("size", "max"),
    zipped_size=("size", "min"),
  )
  .reset_index()
)
gdf["compression_ratio"] = gdf["unzipped_size"] / gdf["zipped_size"]

gdf
```
